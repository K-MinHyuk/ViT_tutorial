```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

import numpy as np
from einops import rearrange

ims = torch.Tensor(np.load('./resources/test_images.npy', allow_pickle=False))
ims = rearrange(ims, 'b h w c -> b c h w')

P = 16
t_img = rearrange(ims[0], 'c h w -> h w c')
H = t_img.shape[0]
W = t_img.shape[1]
N = int((H*W)/(P**2))

fig = plt.figure(figsize=(10,6))
plt.imshow(t_img)
plt.clim([0,1])
plt.hlines(np.arange(P, H, P)-0.5, -0.5, W-0.5, color='w')
plt.vlines(np.arange(P, W, P)-0.5, -0.5, H-0.5, color='w')
plt.xticks(np.arange(-0.5, W+1, 10), labels=np.arange(0, W+1, 10))
plt.yticks(np.arange(-0.5, H+1, 10), labels=np.arange(0, H+1, 10))
x_text = np.tile(np.arange(8, W, P), 6)
y_text = np.repeat(np.arange(8, H, P), 6)
for i in range(1, N+1):
    plt.text(x_text[i-1], y_text[i-1], str(i), color='b', fontsize='xx-large', ha='center')
```
